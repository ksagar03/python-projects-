```python
def Euclidean_Distance_find(point1, point2):
    distance = 0
    if len(point1) == len(point2):
        for i in range(len(point1)-1):
            distance += (point1[i] - point2[i])**2
    return distance**0.5


def Euclidean_Distance(point1, point2):
    if len(point1) == len(point2):
        return np.sqrt(np.sum(((point1 - point2)[:-1])**2))
        
```

```python
def Get_NN(Train, TestRow, num):
    """
    1. Find Dis of TestRow with All Train Data,
    2. Sort Our Data acc to Dis.
    3. Slice Data based on Number
    4. Find MaX Class...
    """
    Distance = []
    data = []
    for i in Train:
        dis = Euclidean_Distance(i, TestRow)
        Distance.append(dis)
        data.append(i)
    Distance = np.array(Distance)
    data = np.array(data)
    ## Find Index of Min Dis... (Sort)
    index_dis = Distance.argsort()
    ## Arange our Data Based on Index_value
    data = data[index_dis]
    ## Slicing Our Number
    nn = data[:num]
    return nn
```

```python
def Predict(Train, NewData, Num):
    Nn = Get_NN(Train, NewData, Num)
    Classes = [i[-1] for i in Nn]
    pred = max(Classes, key=Classes.count)
    return np.array(pred, dtype=int)
    
```

### Create Algo

```python
import numpy as np
class KNClassifier:
    neighbors = 0
    my_data = ""
    def __init__(self, n_neighbors = 3):
        self.neighbors = n_neighbors
        
    def fit(self, Ind, Dep):
        if len(Ind) != len(Dep):
            raise TypeError("Shape is not Same. ")
        
        n_col = Ind.shape[1]
        complate_data = np.insert(Ind, n_col, Dep, axis = 1)
        sf = np.random.permutation(len(Ind))
        self.my_data = complate_data
        
        
    def Euclidean_Distance(self, point1, point2):
        point1 = point1[:-1]
        if len(point1) == len(point2):
            return np.sqrt(np.sum(((point1 - point2))**2))
        else:
            print(-1)
    
        
    def Get_NN(self, Train, TestRow, num):
        """
        1. Find Dis of TestRow with All Train Data,
        2. Sort Our Data acc to Dis.
        3. Slice Data based on Number
        4. Find MaX Class...
        """
        Distance = []
        data = []
        for i in Train:
            dis = self.Euclidean_Distance(i, TestRow)
            Distance.append(dis)
            data.append(i)
        Distance = np.array(Distance)
        data = np.array(data)
        ## Find Index of Min Dis... (Sort)
        index_dis = Distance.argsort()
        ## Arange our Data Based on Index_value
        data = data[index_dis]
        ## Slicing Our Number
        nn = data[:num]
        return nn

    def predict(self, NewData):
        all_pred = []
        for i in NewData:
            Nn = self.Get_NN(self.my_data, i, self.neighbors)
            Classes = [i[-1] for i in Nn]
            pred = max(Classes, key=Classes.count)
            all_pred.append(np.array(pred, dtype=int))
        return np.array(all_pred) 
    
    def score(self, act, pred):
        n_correct = 0
        for i in range(len(act)):
            if act[i] == pred[i]:
                n_correct = n_correct + 1
        acc = n_correct/(len(act))
        return acc
        
```

```python
model = KNClassifier(n_neighbors=12)
```

```python
model.fit(data, target)
```

```python
pred = model.predict(data)
```

```python
model.score(pred, target)
```
